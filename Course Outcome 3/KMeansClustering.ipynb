```python
import pandas as pd
```

```python
ds=pd.read_csv('data.csv')
ds
```

```python
from sklearn.cluster import KMeans
```

```python
c=KMeans(n_clusters=2)
```

```python
c.fit(ds.drop('private',axis=1))
```

```python
c.cluster_centers_
```

```python
def fun(data):
    if data=="Yes":
        return 1
    else:
        return 0
```

```python
ds['cluster']=ds['private'].apply(fun)
```

```python
ds
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print(confusion_matrix(ds['cluster'],c.labels_))
```

```python
print(classification_report(ds['cluster'],c.labels_))
```

```python
accuracy_score(ds['cluster'],c.labels_)
```
