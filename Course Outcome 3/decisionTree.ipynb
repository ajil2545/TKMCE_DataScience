```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# Loading the dataset
iris = load_iris()

# Forming the iris dataframe
# X=iris.data
X = iris.data

Y=iris.target
print("\nClass Labels for all the data points:\n", Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,random_state=20,test_size=0.3)
```

```python
# Defining Decision Tree Algorithm

# Fit the classifier with default hyper-parameters
dtree = DecisionTreeClassifier()   # (random_state=1234)
model=dtree.fit(Xtrain,Ytrain)
```

```python
fig = plt.figure(figsize=(25,20))
dt = tree.plot_tree(dtree, rounded=True,
                   feature_names=iris.feature_names,  
                   class_names=iris.target_names,
                   filled=True)
```

```python
from sklearn.metrics import accuracy_score
```

```python
p1=dtree.predict(Xtest)
```

```python
accuracy_score(Ytest,p1)
```

```python
dtree2 = DecisionTreeClassifier(criterion='entropy')   # (random_state=1234)
model2=dtree2.fit(Xtrain,Ytrain)
```

```python
fig = plt.figure(figsize=(25,20))
dt2 = tree.plot_tree(dtree2,rounded=True,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
```

```python
p2=dtree2.predict(Xtest)
```

```python
accuracy_score(Ytest,p2)
```
